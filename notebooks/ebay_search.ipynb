```python
import os
import sys
from pathlib import Path

# Add the project root to the Python path
project_root = Path().resolve().parent
sys.path.insert(0, str(project_root))

import pandas as pd
from dotenv import load_dotenv

pd.set_option("display.max_colwidth", None)

load_dotenv()

from src.tools.ebay_search import EbaySearch
```

## Initialize eBay Search

```python
# Initialize the EbaySearch tool
ebay_search = EbaySearch()
```

## Example Search

Search for a Pokémon card with a specific condition.

```python
# Define card data to search for
card_data = {
    "name": "Japanese Pikachu Gold Star",  # Card name
    "condition": "PSA 10",  # Card condition (e.g., NM, LP, MP, HP, DMG)
}
```

```python
# Perform the search
results_df = ebay_search.search(card_data)
print(f"\nFound {len(results_df)} results")
```
